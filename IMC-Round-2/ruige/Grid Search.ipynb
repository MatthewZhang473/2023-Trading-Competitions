```python
import os

# iterate over all the files in the directory
for filename in os.listdir(grid_dir):
    if filename.endswith('.txt'):
        # extract the value of x from the filename
        x = int(filename.split('_')[-1].split('.')[0])
        print(x)
        # open the file and read its contents
        with open(os.path.join(grid_dir, filename), 'r') as f:
            contents = f.read()

        # replace the line containing 'big_diff=' with 'big_diff=x'
        contents = contents.replace('big_diff = 30', f'big_diff={x}')
        # write the modified contents back to the file
        with open(os.path.join(grid_dir, filename), 'w') as f:
            f.write(contents)
```
