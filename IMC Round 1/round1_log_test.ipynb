### Parse and Plot Profits From Log

```python
import re

log_file_path = "mm_minexit_lot=5_pos_control.log"

with open(log_file_path, "r") as f:
    log_file = f.read()

profits = re.findall(r"if it is last round, profit = (-?\d+(?:\.\d+)?)", log_file)
positions = [0,0]+ re.findall(r"position now is {'BANANAS': (-?\d+(?:\.\d+)?)", log_file)
```

```python
import matplotlib.pyplot as plt

length = 800
x = range(length)
y = [float(num) for num in profits]
pos = [float(num) for num in positions]

fig, (ax,ax2) = plt.subplots(2,1,figsize=(12, 6),sharex=True)
ax2.plot(x, pos[:length])
ax.plot(x, y[:length])
plt.show()
```
